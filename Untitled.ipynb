```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import new_lib as nl
import sklearn.preprocessing
from sklearn.model_selection import train_test_split
import wrangle as w
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
from regprepare import get_auto_mpg, train_val_test
import warnings
warnings.filterwarnings("ignore")
from math import sqrt
```

```python
zil = w.wrangle_zillow()
zil
# acquiring and prepping the data set 
```

```python
zil, fences = nl.remove_outliers(zil)
```

```python
zil
```

```python
train, val, test = train_val_test(zil, stratify = 'tax_value')
train.shape, val.shape, test.shape
```

```python
preds = train[['square_footage', 'tax_value']]
```

```python
preds['baseline'] = round(train['tax_value'].mean(), 2)
preds.head()
```

```python
ols_model = LinearRegression()

ols_model.fit(train[['square_footage']], train[['tax_value']])
```

```python
preds['y_hat'] = ols_model.predict(train[['square_footage']])
```

```python
preds['baseline_residuals'] = preds.baseline - preds.tax_value
preds['residuals'] = preds.y_hat - preds.tax_value
```

```python
preds
```

```python
plt.scatter(preds.square_footage, preds.baseline_residuals)
plt.xlabel('Square Footage')
plt.ylabel('Baseline Residual')
plt.show()
```

```python
plt.scatter(preds.square_footage, preds.residuals)
plt.xlabel('Square Footage')
plt.ylabel('Residual')
plt.show()
```

```python
preds['baseline_res_squared'] = preds['baseline_residuals'] ** 2
preds['residuals_squared'] = preds['residuals'] ** 2
preds
```

```python
b_SSE = preds.baseline_res_squared.sum()
p_SSE = sum(preds.residuals_squared)
b_SSE, p_SSE
```

```python
b_MSE = b_SSE/len(preds)
p_MSE = p_SSE/len(preds)
b_MSE, p_MSE
```

```python
b_RMSE = sqrt(b_MSE)
p_RMSE = sqrt(p_MSE)
b_RMSE, p_RMSE
```

```python
b_ESS = ((preds.baseline - preds.baseline) ** 2).sum()
p_ESS = ((preds.y_hat - preds.baseline) ** 2).sum()
b_ESS, p_ESS
```

```python
b_TSS = b_ESS + b_SSE
p_TSS = p_ESS + p_SSE
b_TSS, p_TSS
```

```python
p_R = p_ESS/p_TSS
p_R
```
